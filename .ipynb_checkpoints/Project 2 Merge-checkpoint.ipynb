```python
import pandas as pd
```

```python
# Import Box Office and Rotten Tomatoes csv

box_office = pd.read_csv("2019-2021 Box Office.csv")
rotten_tomatoes = pd.read_csv("project-2/RT Ranking 2019-2021.csv")

# Change column header for movie title to match

box_office = box_office.rename(columns = {"Movie":"Title"})

```

```python
# Merge on Title
box_office_rating_merge = pd.merge(box_office,rotten_tomatoes,on = "Title",how = "left")
box_office_rating_merge = box_office_rating_merge.rename(columns = {"Rank_x":"Box Office Rank","Rank_y":"Rotten Tomatoes Rank"})
box_office_rating_merge
```
