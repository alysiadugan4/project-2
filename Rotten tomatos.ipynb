```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url2021 = 'https://www.rottentomatoes.com/top/bestofrt/?year=2021'
url2020 = 'https://www.rottentomatoes.com/top/bestofrt/?year=2020'
url2019 = 'https://www.rottentomatoes.com/top/bestofrt/?year=2019'
# browser.visit(url)

```

```python
import ssl
import certifi
import urllib

response1 = urllib.request.urlopen(url2021, context=ssl.create_default_context(cafile=certifi.where()))
response2 = urllib.request.urlopen(url2020, context=ssl.create_default_context(cafile=certifi.where()))
response3 = urllib.request.urlopen(url2019, context=ssl.create_default_context(cafile=certifi.where()))
```

```python
df2021 = pd.read_html(response1)
df2020 = pd.read_html(response2)
df2019 = pd.read_html(response3)
```

```python
R2021movies = df2021[2]
# df['result'] = df['result']. str. replace(r'\D', '')
R2021movies['Title']= R2021movies['Title'].replace("(2021)","")
R2021movies['Title'] = R2021movies['Title'].str.replace(r" \(.*\)","")
```

```python
R2020movies = df2020[2]
R2020movies['Title']= R2020movies['Title'].replace("(2020)","")
R2020movies['Title'] = R2020movies['Title'].str.replace(r" \(.*\)","")
R2020movies
```

```python
R2019movies = df2019[2]
R2019movies['Title']= R2019movies['Title'].replace("(2019)","")
R2019movies['Title'] = R2019movies['Title'].str.replace(r" \(.*\)","")
R2019movies
```

```python
R2021movies.to_csv("R2021movies.csv", index=True)
```
